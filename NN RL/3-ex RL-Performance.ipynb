```python
from environment import NormalMoveEnv
from experiment import Trajectory
from nn import BetaModel, Model
from rl import Agent

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# %matplotlib inline
```

# Env

```python
env = NormalMoveEnv()

s = env.reset()
print(s)
print(env.step(0))

# ax = env.plot()
```

```python
n_models = 5

agents = [Agent(env) for _ in range(n_models)]

sums = []
for a in agents:
    a.train(model_learn=False)
    sums = [t.run.sum().r for t in a.trajectories]

_, ax = plt.subplots(1, 1, figsize=(10, 4))
ax.plot([s/n_models for s in sums])
```

```python
a = Agent(env, state_space_size=40)
a.train(model_learn=False, episodes=1000)
a.plot()
```
