```python
from bandit import Bandit, BanditExperiment
import matplotlib.pyplot as plt
from plots import testbed

b1 = Bandit(e=0.1)
experiment_1 = BanditExperiment(bandits_config=b1.config, runs=1)
experiment_1.getResults()

b2 = Bandit(e=0.5)
experiment_2 = BanditExperiment(bandits_config=b2.config, runs=1)
experiment_2.getResults()

b3 = Bandit(e=0.9)
experiment_3 = BanditExperiment(bandits_config=b3.config, runs=1)
experiment_3.getResults()

fig, axes = plt.subplots(3, 1, figsize=(16, 16))
testbed(experiment_1.results, axes[0], 'Testbed sample average e=0.1')
testbed(experiment_2.results, axes[1], 'Testbed sample average e=0.5')
testbed(experiment_3.results, axes[2], 'Testbed sample average e=0.9')
```

```python
from bandit import Bandit, BanditExperiment
import matplotlib.pyplot as plt
from plots import testbed
from plots import avgPerformance

b1 = Bandit(e=0.1)
b2 = Bandit(e=0.01)
experiment_4 = BanditExperiment(bandits_config=[b1.config, b2.config], steps=500, runs=100)
result_4, bandits = experiment_4.getResults()

fig, axes = plt.subplots(2,1, figsize=(16, 12))
avgPerformance(result_4[result_4.bandit == 0], axes, 'e=0.1')
avgPerformance(result_4[result_4.bandit == 1], axes, 'e=0.01')
```
